# Polynominal Regression 
## Introducation:
Polynominal regression is a type of regression analysis in which the relationship between independent variable x and dependent variable y is modeled as an nth degree polynominal.\
Polynominal regression fits a  nonlinear relationship between the value of x and crossponding conditional mean of y,denoted as $$E(y|x)$$ and has been used to describe the non linear phenomena such as the growth rate of tissues, distribution of carbon isotops in lake sediments and progression of the disease epidemics.

## Difference between Linear, MultiLinear and polynominal Regression 
### Mathmatical equation of linear Regression
$$ y = b0 +b1x1 $$ 
### Mathmatical equation of Multilinear Regression
$$ y = b0 + b1x1 + b2x2 +........ +bnxn $$
### Polynominal Regression 
$$ y= b0 + b1x1 +b2x2^2 +.....+bnxn^n $$


## keys points of polynomial regression 
### 1. Flexibility in Modeling Curves:
Unlike linear regression, polynominal regression can model non linear relationship that are not straight lines.So it can be very useful in modeling curves.
### 2. Choice of Degree:
A degree of choice is a crucial decision. A polynominal too high in degree can lead to overfiting. where model sarts to model the random noise in the data.
### 3. Still Linear Model
Despite dealing with non linear relationships, polynominal regression is still considered a linear model. This is because its linear in its coefficient Bi, not in featues.
### 4. Potential for Extrapolation Errors
High degree polynominal can lead to extreme values outside the range of the data, leading to poor Extraploation performance.

### EXample of Polynominal Regression

```python
# importing libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
# machine learning 
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import operator
# Step 1: Creat a dummy data
np.random.seed(0)
x = 2-3* np.random.normal(0,1,20)
y = x-2 * (x**2)+np.random.normal(-3,3,20)
# Reshaping data for the model 
x = x[:,np.newaxis]
y = y[:,np.newaxis]
# Step 2: Polynominal transformation
degree = 2 # degree of polynomial 
poly_features = PolynomialFeatures(degree=degree)
x_poly= poly_features.fit_transform(x)
# Step 3: Fit the model 
model = LinearRegression()
model.fit(x_poly,y)
y_poly_pred= model.predict(x_poly)
# Step 4: Ploting the results
plt.scatter(x,y,color='blue')
sorted_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,y_poly_pred),key=sorted_axis)
x, y_poly_pred= zip(*sorted_zip)
plt.plot(x,y_poly_pred,color='magenta')
plt.title('Polynominal Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
