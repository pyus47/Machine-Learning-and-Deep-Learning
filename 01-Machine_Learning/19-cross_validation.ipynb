# Cross validation
- cross validation is a set of techniques of  evaluating a model by spliting the data into a training and test data.
- The traing set is used to train the model, while testing set is used to evaluate the model's performance
- Cross validation is a statistical techniques used in machine learning and data analysis to evaluate how well a model will be able to generlize on the new data.
- In a prediction problem, a mode is usually given a training data set in order to learn pattern in it,and given an unkwon dataset(first seen dataet) aganist which a model is tested.(called the testing or validation dataset)\
**`The basic idea is to divide the dataset into two parts`**
1. A training set, which is used to train a model 
2. A testing set or valdiation set which is used to evaluate performance of model.
### K-fold cross validation
In a k-fold cross validation aviable dataset is divided in k equal parts or folds. The model is then tranied on k-1 fold and validated on the remaning fold. This processs begin repeated a k times, with each fold beings used once as a validation set. The result from each fold is then averaged in order to obtain the overall estimate performance of the model.
## Types of Cross validation 
1. K-fold validation
2. train_test_split
3. leave-one-out-cross-validation
4. straified cross validation
5. Times series cross validation 
6. Group cross validation

```python
import seaborn as sns 
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
# load the iris data set 
import matplotlib.pyplot as plt
iris= load_iris()
nb=GaussianNB()
scores= cross_val_score(nb,iris.data,iris.target,cv=5, scoring='accuracy')
# print the scores for each fold and mean score 
print("Scores for each fold:", scores)
print("Mean scores:",scores.mean())
print('Standard deviation:',scores.std())
```

# K-fold Cross validation vs Train_test_split
#### 1. Data usage 
In a k-fold cross valdiation, the data is split into k equap parts or folds. The model is trained on k-1 fold and then tested on the remaining fold. This process is used k times with each fold being used once as a valdiation set.\
In contrast, train_test_split divides the data into two parts training dataset and testing dataset. In typically the ratio is 70,30, or 80 ,20 ratio of traning and testing data.
#### Data Size
K-fold validation is often used when the data size is relatively small as it  allows for the better use of the data 
In contrast, train_test_split is used when data set is large.As it is faster to implement and may be sufficient to evaluating the model performance
##### Performance estimation
K-fold valdiation provides better model performance estimation as it is used on mutliple independent subset of the data. This help to detect the variance and overfiting of the model.
In contrast, train_test_split provides less accurate estimation on the model's performance. As it is used on the specific subset of the data 
#### computation time
k-fold cross valdiation can computationally expensive as it requires the training and valdiating the model k times.
In contrast train_test_split is faster as it requires training and validating the model only once.

#### summary 
In summary k-fold cross validation is more robust and accurate technique of model performance evluation as compared to when the dataset is relatively small.
Train test split is a faster and simpler technique that can be used. when the dataset is larger and quick estimation of performance is needed.

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
import seaborn as sns 
# load the data 
df = sns.load_dataset('tips')
x = df[['size', 'total_bill','tip']]
y= df['sex']
# define the Gusasain 
nb= GaussianNB()
# 
scores= cross_val_score(nb,x,y,cv=4,scoring='accuracy')
# print scores for each fold and mean score 
print("scores of each fold",scores)
print('mean scores', scores.mean()
)
print("standard deviation", scores.std())
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
iris= load_iris()
x = iris.data
y = iris.target
clf = DecisionTreeClassifier()
scores= cross_val_score(clf,x,y ,cv=8,scoring='accuracy')
# plot scores of each fold 
plt.figure(figsize=(8,5))
plt.plot(range(1,9),scores,color='blue',linestyle='dashed',marker='o',markeredgecolor='red',markeredgewidth=2.5)
plt.title('scores of  each fold  iris data ')
plt.xlabel('fold number')
plt.ylabel('accuracy')
plt.ylim(0.5,1.0)
plt.show
```
