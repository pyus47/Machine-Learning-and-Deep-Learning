# Inverse Encode / Inverse Transform / Decdoing the categorical columns 
Inverse Encoding is a process of decoding categorical data back to its original form.This is done by using a method called inverse_transform.This method is avialable with following enocders.
1. One Hot Encoder
2. LabelEncoder
3. OrdinalEncoder

```python
# import libaries 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,OrdinalEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
# load the dataset 
df = sns.load_dataset('titanic')
```

```python
# droping the deck column
df.drop('deck',axis=1,inplace=True)
```

```python
# imputing missing val
df['age'].fillna(df['age'].median(),inplace=True)
df['embark_town'].fillna(df['embark_town'].mode()[0],inplace=True) 
df['embarked'].fillna(df['embarked'].mode()[0],inplace=True)
```

### Label Encoder

```python

# we have to make a seperate encoder for each column
le_sex =LabelEncoder()
le_who = LabelEncoder()
le_class= LabelEncoder()
le_embark_town= LabelEncoder()
le_embarked = LabelEncoder()

# now fit these encoders on to the data 
df['sex']= le_sex.fit_transform(df['sex'])
df['who']=le_who.fit_transform(df['who'])
df['class']=le_class.fit_transform(df['class'])
df['embark_town']=le_embark_town.fit_transform(df['embark_town'])
df['embarked']= le_embarked.fit_transform(df['embarked'])
```

```python
df.head()
```

```python
df['sex']=le_sex.inverse_transform(df['sex']) 
df['embarked']= le_embarked.inverse_transform(df['embarked'])
df['embark_town']= le_embark_town.inverse_transform(df['embark_town'])
df['class']= le_class.inverse_transform(df['class'])
df['who']= le_who.inverse_transform(df['who'])
```

```python
df.head()
```

### One Hot Encoder

```python
# Now using one hot encoder to encode the values 
ohe_sex = OneHotEncoder()
ohe_who= OneHotEncoder()
ohe_class= OneHotEncoder()
ohe_embark= OneHotEncoder()
ohe_embarked= OneHotEncoder()
# now fit these encoders on the data 
df['sex']= ohe_sex.fit_transform(df[['sex']]).toarray()
df['class']= ohe_class.fit_transform(df[['class']]).toarray()
df['embark_town']= ohe_embark.fit_transform(df[['embark_town']]).toarray()
df['embarked']= ohe_embarked.fit_transform(df[['embarked']]).toarray()
df['who']= ohe_who.fit_transform(df[['who']]).toarray()
df.head()
```

```python
df['sex']= ohe_sex.inverse_transform(df['sex'])
```

### Ordinal Encoder 

```python
od_sex = OrdinalEncoder()
df['sex']= od_sex.fit_transform(df[['sex']])
df.head()
```

```python
df['sex']= od_sex.inverse_transform(df[['sex']])
df.head()
```

```python
od_who = OrdinalEncoder()
df['who']= od_who.fit_transform(df[['who']])
df.head()
```

```python
df[['who']]= od_who.inverse_transform(df[['who']])
```
