```python
# importing libraries 
import yfinance as yf 
import pandas as pd

# define tikcer symbol
tickerSymbol = 'AAPL'

# get data on this ticker

tickerData= yf.Ticker(tickerSymbol)
# information lof ticker data
tickerData.info
```

```python
# get the historical prices for this ticker 
tickerDf = tickerData.history(period='1d',start='2010-1-1',end='2020-1-25')
tickerDf['Close'].iloc[-1]

# print head of the data 
print(tickerDf.head())

print(tickerDf.shape)
```

```python
# importing libraries

import pandas as pd 
import yfinance as yf 
import datetime as dt 
from datetime import date, timedelta

today = date.today()
d1 = today.strftime('%Y-%m-%d')
d2 =(today-timedelta(days=365)).strftime('%Y-%m-%d')

# getting data from Yahoo Finance for
start_date= d2
end_Date=d1

# tikcer 

tickerSymbol = 'TSLA'

# get data on this ticker 
tickerData=yf.Ticker(tickerSymbol)

# get the historical prices of this ticker

tickerDf = tickerData.history(period='1d',start=start_date,end=end_Date)

# last closing prices 
print(tickerDf.head())
print(tickerDf.shape)
```
