```python
# importing libraries 
import pandas as pd 
from  bs4 import BeautifulSoup
import requests

# getting website url 

url = "https://en.wikipedia.org/wiki/World_population"
# accessing website through  'requests' package and saving it in a variable called response

response = requests.get(url)

# now using  the 'BeautifulSoup' library to parse the data from the webpage
soup = BeautifulSoup(response.content, 'html.parser')  

# printing the title 
print(soup.title)

# priting without html tags
print(soup.title.text)

# getting html elements 

# print(soup.prettify())
# printing the text of webpage
# print(soup.text)

# find all the tables 

tables = soup.find_all('table')

# print(tables)

# making a dataframe of all the tables in the page
dataframe = []

for i, table  in enumerate(tables):
    rows = table.find_all("tr")[1:]
    data = []
    for row in rows:
        cols = row.find_all(['th','td'])
        col = [col.text.strip() for col in cols]
        data.append([col])
        df = pd.DataFrame(data)
    dataframe.append(df)

dataframe[4]
    

```
