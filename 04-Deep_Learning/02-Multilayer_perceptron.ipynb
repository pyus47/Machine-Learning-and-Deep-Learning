# Multilayer Perceptron:
A Multilayer Perceptron is a type of a neural network that consist of  a more then one layer of neurons. Unlike a single layer Perceptron, which only learns linearly seperable patterns, a multilayer can learn more complex non-linear functions. This makes it  a fundamental model in field of deep learning and neural networks.
## Structure of Multilayer Perceptron:
### 1. input layer:
That Recives the input data. These input are passed to the  next layers.
### 2. Hidden Layers 
one or more hidden layers, where actual processing is done through the  system  of weigthed connections. The number of of hidden layers and  number of neurons are  adjustable parameters and can significantly influence the model ability to capture complex patterns in the data.
### 3. Output Layer: 
This layer produce the final output. The design of the output layer depend upon on specific task (e.g. regression, classification task)
## How it works ? 
### 1. Forwad propagation:
Singal travesl from input layer to the final output layers. Each neuron in the hidden layers and output layers processes the input signal using an acitvation function, which introduce non-linear properties to the network.
### 2. Back propagaion:
After the ouptut is generated, The Mlp uses a method called backporpagation to update the weight of the neurons. During the backporpagation the error (the difference between the actual and predicated ) is calculated and distributed back through the network, allowing the weights to updated accordingly.
### 3. Learning Rate:
This is a key parameter in training an Mlp. it controls how much weights are adjusted during backprogation.
### 4. Activation Function: 
Function like sigmoid, Relu (Rectified Linear Unit), or Tanh(Hyperbolic Tangnet) are used to non-linear properties, allowing The MlP to learn more complex patterns.
## Types of Multilayer Precptron:
#### 1. Based on the hidden layer

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
# load the dataset 
tips = sns.load_dataset('tips')
```

```python
# preprocessing of the data 
import tensorflow as tf

# converting categorical data into numeric one 
tips = pd.get_dummies(tips,drop_first=True)
tips.head()
```

```python
# selecting features and target variable
x = tips.drop('tip',axis=1)
y = tips['tip']
# spilting the dataset into training and testing dataset 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
# standarizing the data or scalling 
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test= scaler.fit_transform(x_test)
# Building the model 
model = tf.keras.models.Sequential
([tf.keras.layers.Dense(64,activation='relu',input_shape=(x_train.shape[1])), # input Layer
                                    tf.keras.Layers.Dense(32,activation='relu'),  # hidden layer
                                    tf.keras.Layers.Dense(1)]) # output layer                 
```

```python
# Building the model 
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(x_train.shape[1])),  # input layer
    tf.keras.layers.Dense(32, activation='relu'),  # hidden layer
    tf.keras.layers.Dense(1)  # output layer
])
# output layer  
```

```python
# Building the model 
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(x_train.shape[1],)),  # input layer
                                             # drop
    tf.keras.layers.Dense(32, activation='relu'),  # hidden layer
    tf.keras.layers.Dense(1)  # output layer
])
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
# Training the model
history = model.fit(x_train, y_train, epochs=20,batch_size=32, validation_data=(x_test, y_test))
# what next
print('Model built and trained!')
# Evaluating the model on test data:
import matplotlib.pyplot as plt

# Evaluate the model on the test data
test_loss, test_mae = model.evaluate(x_test, y_test)
print(f'Test Loss: {test_loss}, Test MAE: {test_mae}')

# Plot training history
history_dict = history.history

# Plot training loss
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(history_dict['loss'], label='Training Loss')
plt.plot(history_dict['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Plot training MAE
plt.subplot(1, 2, 2)
plt.plot(history_dict['mae'], label='Training MAE')
plt.plot(history_dict['val_mae'], label='Validation MAE')
plt.title('Training and Validation MAE')
plt.xlabel('Epochs')
plt.ylabel('MAE')
plt.legend()

plt.show()


         
```
