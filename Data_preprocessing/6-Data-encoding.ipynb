# Data encoding
1. converting categorical column into numerical type 
### types of data encoding
1. `label encoding`give each categorical value to crossponding numerical value\
 (for example=['usama','hamza','basit','iqra']==[1,2,3,4])
2. `one hot encoding` we can make a separte coloum for each`table for each` only one value which is going to true in one row.\
|usama| hamza| basit| iqra|\
|0    |1     | 0    |0 
3. `ordinal encoding `. in this encoding an order based to some value to each categorical value\
for example usama is bigess he will 1,basit,2 iqra 3,hamza 4
4. `binary ecoding` we merge these four column into two column only.
5. `frequency and count encoding` inseated of writing their names we can write their frequency count number.\
for example hamza has been mentioned in 23, usama is 45,basit is 67, (23,45,67)

#### why we need to encode data 
1. `algorithim campatibilty `(algorithim are more comfortable working numeric data rather then text data )
2. `efficiency and performance`( faster and storage efficient)
3. `feature representation` ( we have reduced biasness of categorical values for example shoe can be written in different languages in udru it will be jota. if we encode this into one it will 1 for shoe 2 for bat)
4. `support the unseen category`(category you werenot seeing earlier now you have to look at them)

## feature encoding
Feature encoding is a processs of transforming categorical features into numerical features.This is necessary because machine learning only handle numeric feature.There are different ways to encode numerical features, and each method has it's own advantage and disadvantages.In this notebook we will explore some of the most popular method for  encoding categorical features.

- label encoding
- one hot encoding
- binary encoding
- ordinal encoding

```python
# let's encode time column with label encoder from sklearn
from sklearn.preprocessing import LabelEncoder ,OneHotEncoder, OrdinalEncoder
# here we are going to label encoder here 
le = LabelEncoder()
df['encoded_time']= le.fit_transform(df[['time']])
df.head()
```

```python
# using ordianl encoder on day column on specific order
# here we are calling ordinal encoder 
oe= OrdinalEncoder(categories=[['Thur','Fri','Sat','Sun']])
# here we are going fit this encoder on data 
df['encoded_day']=oe.fit_transform(df[['day']])
df.head()
```

```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
pd.set_option('display.max_columns', None)  # To display all columns

pd.set_option('display.width', None)

# or
pd.set_option('display.max_columns', 30)    # To display a specific number of columns

# Load the Titanic dataset
df = sns.load_dataset('titanic')

# Create an instance of OneHotEncoder
one = OneHotEncoder()

# Fit and transform the 'embarked' column and convert it to a DataFrame
one_df = pd.DataFrame(one.fit_transform(df[['embarked']]).toarray(), columns=one.get_feature_names_out(['embarked']))

# Concatenate the original DataFrame with the one-hot encoded DataFrame
df = pd.concat([df, one_df], axis=1)

# Drop the original 'embarked' column as it's no longer needed
df = df.drop(['embarked'], axis=1)

# Display the modified DataFrame
print(df.head())
```
