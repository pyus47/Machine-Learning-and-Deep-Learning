# Hirerachial Clustering


```python
# An example of Hirerachial clustering in notebook
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch 
from sklearn.datasets import make_blobs 
# making an datasets 
x, y = make_blobs(n_samples=100,centers=3,n_features=3,random_state=42)
# perform hirerachial clustering using agglomerativeclustering
clustering = AgglomerativeClustering(n_clusters=8,linkage='ward',metric='euclidean')
(clustering.fit(x))
print(clustering.labels_)
print(y)
# plot the dendogram
plt.figure(figsize=(10,7))
plt.title('Dendrogram')
dendrogram= sch.dendrogram(sch.linkage(x,method='ward'))
plt.show()
```

```python
# import libraries 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn  as sns 
import pandas as pd 
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch 
# load the dataset 
df = sns.load_dataset('tips')
# making clustering data 
x = df[['total_bill','tip']]
# calling model here 
clustering = AgglomerativeClustering(n_clusters=3)
# making predication from fitting data 
agg_labels = clustering.fit_predict(x)
df['agg_labels']=agg_labels
# now making scatterplot to see clusters 
plt.figure(figsize=(10,6))
plt.title('Agglomerative Clustering on tips Dataset')
plt.xlabel('total bill')
plt.ylabel('tip')
sns.scatterplot(data=df,x='total_bill',y='tip',hue='agg_labels',palette='viridis')
plt.show()
# making dendrogram here 
plt.figure(figsize=(10,7))
plt.title('Dendrogram')
dendrogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.show()
```
